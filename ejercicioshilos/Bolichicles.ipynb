```
import threading
from threading import Thread, RLock
from queue import Queue
from random import randint
import math
import time   
import sys

class ProdChicles(Thread):
    def __init__(self,mutex):
        super().__init__()
        self.mutex = mutex

    def run(self):
      global cola_chicles
      while True:
        self.mutex.bloquear()
        cola_chicles.put(1)  
        sys.stdout.write(f"Se ha creado un chicle\n") 
        self.mutex.desbloquear()
        time.sleep(1)

class ProdPolvos(Thread):
    def __init__(self,mutex):
        super().__init__()
        self.mutex = mutex

    def run(self):
      global cola_polvos
      while True:
        self.mutex.bloquear()
        cola_polvos.put(1)
        sys.stdout.write(f"\nSe produce un polvo\n")  
        self.mutex.desbloquear()
        time.sleep(1)

class Maquina_cubre(Thread):
    def __init__(self,mutex_chicle,mutex_polvo):
        super(Maquina_cubre,self).__init__()
        self.mutex_chicle = mutex_chicle
        self.mutex_polvo=mutex_polvo

    def run(self):
      global cola_polvos
      global cola_chicles    
      while True:
        self.mutex_chicle.bloquear()
        self.mutex_polvo.bloquear()
        if (cola_polvos.qsize() !=0 and cola_chicles.qsize()!=0):
          cola_polvos.get()
          cola_chicles.get()
          cola_bolichicle.put(1)
          sys.stdout.write(f"\n**Se produce un bolichicle\n")  
        self.mutex_chicle.desbloquear()
        self.mutex_polvo.desbloquear()
        time.sleep(2)

class Maquina_bolsa(Thread):
    def __init__(self,mutex):
        super(Maquina_bolsa,self).__init__()
        self.mutex = mutex

    def run(self):
      global cola_bolichicle
      global cola_bolsa      
      while True:
        self.mutex.bloquear()
        if(cola_bolichicle.qsize() ==10):
            for i in range(10):
                cola_bolichicle.get()
            cola_bolsa.put(1)
            sys.stdout.write(f"\n--->Se ha creado una bolsa\n")  
        self.mutex.desbloquear()

class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.RLock = RLock()
    
    def bloquear(self):
        self.RLock.acquire()
        
    def desbloquear(self):
        self.RLock.release() 

if __name__=="__main__":
  mutex_chicle=Mutex()
  mutex_polvo=Mutex()
  mutex_bolichicle=Mutex()
  cola_chicles=Queue()
  cola_polvos=Queue()
  cola_bolichicle=Queue()
  cola_bolsa=Queue()

  prodChicles=ProdChicles(mutex_chicle)
  prodPolvos=ProdPolvos(mutex_polvo)
  bolichicle=Maquina_cubre(mutex_chicle,mutex_polvo)
  bolsa=Maquina_bolsa(mutex_bolichicle)

  prodChicles.start()
  prodPolvos.start()
  bolichicle.start()
  bolsa.start()

  prodChicles.join()
  prodPolvos.join()
  bolichicle.join()
  bolsa.join()
```
