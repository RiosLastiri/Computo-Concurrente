```
import threading
from threading import Thread, Lock, RLock
import random
import time
import sys

class Tortuga(Thread):
  def __init__(self,inicio,fin,meta,lock):
    super(Tortuga,self).__init__()
    self.fin=fin
    self.inicio=inicio
    self.posicion=1
    self.meta=meta
    self.lock=lock
    self.timeStamp=float('inf')
  
  def run(self):
    while self.posicion<self.fin:
      time.sleep(1)
      self.suceso(random.random())
      sys.stdout.write(f"posicion-{self.posicion} T\n")

    self.timeStamp=time.time()

  def suceso(self,probabilidad):
    if 0<probabilidad<=0.5:
      self.posicion+=3

    elif 0.5<probabilidad<=0.7:
      self.posicion-=6

    elif 0.7<probabilidad<=1.0:
      self.posicion+=1
    
    if self.posicion<1:
      self.posicion=1


class Liebre(Thread):
  def __init__(self,inicio,fin,meta,lock):
    super(Liebre,self).__init__()
    self.fin=fin
    self.inicio=inicio
    self.posicion=1
    self.meta=meta
    self.lock=lock
    self.timeStamp=float('inf')
  
  def run(self):
    while self.posicion<self.fin:
      time.sleep(1)
      self.suceso(random.random())
      sys.stdout.write(f"posicion-{self.posicion} L\n")
      
    self.timeStamp=time.time()

  def suceso(self,probabilidad):
    if 0.2<probabilidad<=0.4:
      self.posicion+=9

    elif 0.4<probabilidad<=0.5:
      self.posicion-=12

    elif 0.5<probabilidad<=0.8:
      self.posicion+=1

    elif 0.8<probabilidad<=1.0:
      self.posicion-=2

    if self.posicion<1:
      self.posicion=1  


if __name__=="__main__":
  meta=[]
  lock=Lock()
  inicio=1
  fin=70
  T=Tortuga(inicio,fin,meta,lock)
  L=Liebre(inicio,fin,meta,lock)

  T.start()
  L.start()

  T.join()
  L.join()

  if L.timeStamp==T.timeStamp:
    sys.stdout.write("Empate")
  elif L.timeStamp<T.timeStamp:
    sys.stdout.write("Gano la Liebre")  
  elif T.timeStamp<L.timeStamp:
    sys.stdout.write("Gano la Tortuga")
```

Producto de dos matrices secuencial

```
import numpy as np
import time
import queue
from queue import Queue

m1 = np.random.randint(0, 100, (100, 100))
m2 = np.random.randint(0, 100, (100, 100))

print("\nM1:\n",m1)
print("\nM2:\n",m2)

def producto(m1,m2):
  if m1.shape[1]==m2.shape[0]:
    resultado=np.zeros((m1.shape[0],m1.shape[0]))
    for i in range(m1.shape[0]):
      for j in range(m2.shape[1]):
        for k in range(m1.shape[1]):
          resultado[i][j] = m1[i][k]*m2[k][j]
    return resultado
  else:
    print('Las matrices no son compatibles')
    return

t=time.time()
mult=producto(m1,m2)
print("\nMultiplicacion secuencial:")
print("M1*M2:\n",mult)
print('\nTiempo total:',time.time()-t)
```
